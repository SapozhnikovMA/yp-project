#### 1.2. Проверка корректности вычисления rougher.output.recovery

```python
# Создали функцию recovery 
def recovery(C, F, T):
    numerator = (C*(F-T))
    denominator =(F*(C-T))
    recovery = numerator / denominator * 100
    return recovery
C,F,T = df_train['rougher.output.concentrate_au'], df_train['rougher.input.feed_au'], df_train['rougher.output.tail_au']
df_train['rougher.output.recovery_count'] = recovery(C, F, T)
df_train[['rougher.output.recovery', 'rougher.output.recovery_count']].head(2)
```

# 3. Создание моделей предсказаний<a id="model"></a>

```python
# Написали функцию sMAPE
def smape(y, y_pred):
    res = abs(y - y_pred) / ((abs(y) + abs(y_pred)) / 2)
    return res.sum() / len(res) * 100
def smape_final(smape_r, smape_f):
    return smape_r * 0.25 + smape_f * 0.75
```

```python
# Подготовим данные для обучения моделей и кросс-валидации
y_rougher = df_train['rougher.output.recovery']
y_final = df_train['final.output.recovery']
X_tr = df_train.drop(['date', 'rougher.output.recovery', 'final.output.recovery'], axis = 1)

features = X_tr.columns
scaler = StandardScaler()
scaler.fit_transform(X_tr[features]) 
X_tr.head()
```

```python
# Создадим константную модель
target_constant_r = pd.Series(y_rougher.median(), index=y_rougher.index)
target_constant_f = pd.Series(y_final.median(), index=y_final.index)
smape_r = smape(y_rougher, target_constant_r)
smape_f = smape(y_final, target_constant_f)
print('Результат константной модели sMAPE (rougher): {:.2f}%'.format(smape_r))
print('Результат константной модели sMAPE (final): {:.2f}%'.format(smape_f))
print('Результат константной модели итоговое sMAPE: {:.2f}%'.format(smape_final(smape_r, smape_f)))
```

```python
# Линейная регрессия
line = LinearRegression()
smape_r = cross_val_score(line, X_tr, y_rougher, cv=5, scoring = make_scorer(smape))
smape_f = cross_val_score(line, X_tr, y_final, cv=5, scoring = make_scorer(smape))
print('Результат модели лин.регрессии sMAPE (rougher): {:.2f}%'.format(smape_r.mean()))
print('Результат модели лин.регрессии sMAPE (final): {:.2f}%'.format(smape_f.mean()))
print('Результат модели лин.регрессии итоговое sMAPE: {:.2f}%'.format(smape_final(smape_r, smape_f).mean()))
```

```python
# DecisionTreeRegressor - подбор параметра
tree = DecisionTreeRegressor()
grid = {'max_depth': range (1, 10)}
tree_rs = RandomizedSearchCV(tree, n_jobs=None, n_iter=10,
                                 param_distributions=grid, random_state=42, scoring=make_scorer(smape, greater_is_better=False))
tree_rs.fit(X_tr, y_rougher)
print('Best Params: \n', tree_rs.best_params_ )
```
